## Check val2017 eng and val2017 vi

```python
import pandas as pd

df_en = pd.read_csv("english/csv/captions_val2017_converted.csv")
df_en.head()
```

```python
df_vi = pd.read_csv("vietnamese/csv/captions_val2017_trans_converted.csv")
df_vi.head()
```

## Merge val2017 vi to val2017 en

```python
df_en["caption_trans"] = df_vi["caption"]
df_en.head()
```

## Merge train2017 vi to train2017 en 

```python
df_train_en = pd.read_csv("english/csv/captions_train2017_converted.csv")
df_train_vi = pd.read_csv("vietnamese/csv/captions_train2017_trans_converted.csv")
df_train_en["caption_trans"] = df_train_vi["caption"]
df_train_en.head()
```

## Merge train2017_en to val2017_en

```python
df_2017_full_en_vi = pd.concat([df_train_en, df_en], ignore_index=True)
df_2017_full_en_vi.head()
```

```python
df_2017_full_en_vi.shape
```

```python
print("Number of unique images in 2017_full_en_vi:", df_2017_full_en_vi['image_id'].nunique())
```

```python
# Export to csv file
df_2017_full_en_vi.to_csv("english/csv/captions_2017_full_en_vi.csv", index=False)
print("Exported to csv file!")
```
