```python
#Count number of images through id
import json

with open('english/json/captions_val2017.json', 'r') as f:
    data = json.load(f)

count_img = 0
count_ann = 0
#Count number of images through id
for image in data['images']:
        count_img += 1
for ann in data['annotations']:
        count_ann += 1

print("Number of images in json file ",count_img)
print("Number of annotations in json file ",count_ann)
```
