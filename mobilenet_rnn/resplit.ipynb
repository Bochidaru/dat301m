Data g·ªëc c√≥ 120k ·∫£nh trong train, 5k ·∫£nh trong val, 40k ·∫£nh test nh∆∞ng m√† ch·ªâ c√≥ captions cho train v√† val, v·∫≠y b·ªè test, t·∫°o test m·ªõi c√≥ 10k ·∫£nh l·∫•y t·ª´ train, v√† c·∫Øt 8k ·∫£nh v√†o val

```python
import os
import random
import shutil
from tqdm import tqdm

# ƒê∆∞·ªùng d·∫´n
train_dir = "coco2017/train2017"
test_dir = "coco2017/test2017"
val_dir = "coco2017/val2017"

# T·∫°o folder n·∫øu ch∆∞a c√≥
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# L·∫•y danh s√°ch to√†n b·ªô ·∫£nh trong train2017
all_images = [f for f in os.listdir(train_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Tr·ªôn ng·∫´u nhi√™n danh s√°ch
random.shuffle(all_images)

# L·∫•y 10k ·∫£nh cho test v√† 8k cho val
test_images = all_images[:10000]
val_images = all_images[10000:18000]

print(f"‚Üí T·ªïng ·∫£nh trong train: {len(all_images)}")
print(f"‚Üí Copy {len(test_images)} ·∫£nh sang test2017")
print(f"‚Üí Copy {len(val_images)} ·∫£nh sang val2017")

# Copy ·∫£nh sang test2017
for img in tqdm(test_images, desc="Copying to test2017"):
    src = os.path.join(train_dir, img)
    dst = os.path.join(test_dir, img)
    shutil.copy2(src, dst)

# Copy ·∫£nh sang val2017
for img in tqdm(val_images, desc="Copying to val2017"):
    src = os.path.join(train_dir, img)
    dst = os.path.join(val_dir, img)
    shutil.copy2(src, dst)

print("‚úÖ Ho√†n t·∫•t chia ·∫£nh!")
```

```python
# G·ªôp l·∫°i th√†nh 1 danh s√°ch c·∫ßn x√≥a
images_to_delete = test_images + val_images
print(f"üóëÔ∏è S·∫Ω x√≥a t·ªïng c·ªông {len(images_to_delete)} ·∫£nh kh·ªèi train2017")

# X√≥a t·ª´ng ·∫£nh
for img in tqdm(images_to_delete, desc="Deleting from train2017"):
    img_path = os.path.join(train_dir, img)
    if os.path.exists(img_path):
        os.remove(img_path)

print("‚úÖ ƒê√£ x√≥a xong c√°c ·∫£nh ƒë√£ copy sang val/test kh·ªèi train2017!")
```

t·∫°o subset

```python
import os, random, shutil
from tqdm import tqdm

base = "coco2017/subset"
os.makedirs(f"{base}/train", exist_ok=True)
os.makedirs(f"{base}/val", exist_ok=True)
os.makedirs(f"{base}/test", exist_ok=True)

source_dir = "coco2017/train2017"

all_images = os.listdir(source_dir)
sample_images = random.sample(all_images, 8000)

# Chia theo t·ª∑ l·ªá 80/10/10
n_total = len(sample_images)
n_train = int(0.8 * n_total)
n_val = int(0.1 * n_total)
n_test = n_total - n_train - n_val  # ph·∫ßn c√≤n l·∫°i

train_imgs = sample_images[:n_train]
val_imgs = sample_images[n_train:n_train + n_val]
test_imgs = sample_images[n_train + n_val:]

def copy_images(img_list, dest):
    for img in tqdm(img_list, desc=f"Copying to {dest}"):
        shutil.copy(os.path.join(source_dir, img),
                    os.path.join(base, dest, img))

copy_images(train_imgs, "train")
copy_images(val_imgs, "val")
copy_images(test_imgs, "test")

print(f"‚úÖ Done: {len(train_imgs)} train, {len(val_imgs)} val, {len(test_imgs)} test images copied.")
```
